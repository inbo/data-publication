```python
def combine_waarnemers(row):
    """ """
    names = row.loc[['WaarnemerNaam1', 'WaarnemerNaam2', 'WaarnemerNaam3']].dropna().values.tolist()
    return " | ".join(names)    
```

```python
data["recordedBy"] = data.apply(combine_waarnemers, axis=1)
```

```python
data["identifiedBy"] = data.apply(combine_waarnemers, axis=1)
```

```python
import requests
```

```python
def get_scientific_name(vernacular_name):
    """ """
    r = requests.get("".join(["http://api.gbif.org/v1/species/search?q=", vernacular_name]))
    try:
        name = r.json()['results'][0]['scientificName']    
        return name
    except:
        return ""
```

```python
data["scientificName"] = data["SoortNaamNl"].apply(get_scientific_name)
```
